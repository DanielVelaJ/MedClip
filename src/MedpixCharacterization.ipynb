```python
import pandas as pd
```

```python
path='../data/intermediate/inter_medpix.csv'
df=pd.read_csv(path)
```

```python
c.index.name='Anatomy'
c=df.groupby(['Anatomy_Group','Modality_Group']).Full_Caption.count().unstack()
c.loc[:,'Total']=c.sum(axis=1)
c.loc['Total']=c.sum(axis=0)
c.index.name=None
c.columns.name='Anatomy'
c=c.sort_values('Total',axis=1).sort_values('Total',axis=0)
index=['Breast',
 'Cardiovascular',
 'Head and Neck',
 'Spine',
 'Genitourinary',
 'Pulmonary',
 'Abdomen',
 'Musculoskeletal',
 'Brain',
 'Other',
 'Total']
c=c.loc[index]
caption='Sample distribution in the MedPix dataset. Columns represent the '+\
        'imaging modalities: HE: Histology, NM: Nuclear Medicine (includin'+\
        'g PET scans), AN: Angiography, US: Ultrasound, XR: Plain X-Ray, CT'+\
        ': Computed Tomography, MR: Magnetic Resonance' 
display(c)
latex_table=c.to_latex(index=True,
                 caption=caption,
                 label='tab: sample distribution',
                 bold_rows=True,
                 column_format='lrrrrrrrr',
                 float_format="%.0f",
                 na_rep='-')
with open('../ECCV_manuscript/tables/'+"Mexpix_characterization.tex", "w") as file1:
    # Writing data to a file
    file1.write(latex_table)
```
